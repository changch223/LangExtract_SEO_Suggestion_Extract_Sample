````python
#Cell 2 — 輔助工具：下載與解析（抓 Title / H2 / H3 / 全文純文字）
import re, requests, bs4
from bs4 import BeautifulSoup
from urllib.parse import urlparse

HEADERS = {
    "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 "
                  "(KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36"
}

def fetch_html(url: str) -> bytes:
    resp = requests.get(url, timeout=30, headers=HEADERS)
    resp.raise_for_status()
    return resp.content  # bytes，交給 BS4 自動偵測 charset

def strip_junk(soup: BeautifulSoup) -> None:
    for bad in soup(["script", "style", "noscript", "svg"]):
        bad.decompose()
    for sel in [
        ".breadcrumbs",".share",".sns",".social",".related",".recommend",
        ".footer",".site-footer",".author",".profile",".tagcloud",
        ".advertisement",".ads",".promo",".newsletter",".cookie",".consent"
    ]:
        for n in soup.select(sel):
            n.decompose()

def extract_title(soup: BeautifulSoup) -> str:
    t = ""
    if soup.title and soup.title.string:
        t = soup.title.string.strip()
    # 也試著抓 h1（有些站 <title> 會帶站名/品牌）
    h1 = soup.find("h1")
    if h1 and h1.get_text(strip=True):
        # 避免太長：優先 h1
        return h1.get_text(" ", strip=True)
    return t

def extract_headings(soup: BeautifulSoup):
    # 只取 h2/h3 的文字做「目次」
    h2s = [h.get_text(" ", strip=True) for h in soup.find_all("h2")]
    h3s = [h.get_text(" ", strip=True) for h in soup.find_all("h3")]
    # 清掉太短或純符號
    def clean_list(xs):
        out = []
        for x in xs:
            x = re.sub(r"\s+", " ", x).strip()
            if len(x) >= 2:
                out.append(x)
        return out
    return clean_list(h2s), clean_list(h3s)

def extract_main_text(soup: BeautifulSoup) -> str:
    # 常見本文容器（按站點可擴充）
    candidates = [
        "article",
        '[role="main"]',
        ".entry-content",
        ".article-body",
        ".post-content",
        ".content__body",
        ".main-contents",
        "#main article",
        ".Article__Body", ".article__content", ".single-content"
    ]
    for sel in candidates:
        node = soup.select_one(sel)
        if node and node.get_text(strip=True):
            text = node.get_text("\n", strip=True)
            break
    else:
        text = soup.get_text("\n", strip=True)

    # 清理多餘空白
    lines = [ln.strip() for ln in text.splitlines()]
    lines = [ln for ln in lines if len(ln) >= 2]
    text = "\n".join(lines)
    text = re.sub(r"\n{3,}", "\n\n", text)
    return text

def load_and_parse(url: str):
    html = fetch_html(url)
    soup = BeautifulSoup(html, "lxml")
    strip_junk(soup)
    title = extract_title(soup)
    h2s, h3s = extract_headings(soup)
    fulltext = extract_main_text(soup)
    return {
        "url": url,
        "title": title,
        "h2": h2s,
        "h3": h3s,
        "text": fulltext,
        "domain": urlparse(url).netloc
    }

# 測試抓第一個（可略）
# data = load_and_parse(TOP_URLS[0]); data["title"], len(data["h2"]), len(data["h3"]), len(data["text"])
````

````python
#Cell 3 — 下載 3 篇文章並彙整
from tqdm import tqdm

docs = []
for u in tqdm(TOP_URLS):
    try:
        docs.append(load_and_parse(u))
    except Exception as e:
        print(f"[WARN] {u} failed: {e}")

assert len(docs) > 0, "No documents parsed."
print("Parsed:", len(docs))
for d in docs:
    print("-", d["url"], "| title:", d["title"][:80])
````

````python
#Cell 4 — Gemini：從「標題 + H2/H3」抽取 競品が使っているサジェスト
#這一步 不會 上傳全文，只用標題與目次（H2/H3）來抽取「exact 出現的片語（2〜12 字）」；之後的出現次數統計會用「全文」本地計算。

import json, textwrap, google.generativeai as genai

genai.configure(api_key=os.environ["GOOGLE_API_KEY"])

def llm_json(model_name: str, system_prompt: str, user_content: str, temperature=0.3):
    model = genai.GenerativeModel(model_name, generation_config={
        "temperature": temperature,
        "response_mime_type": "application/json"
    })
    prompt = system_prompt + "\n\n" + user_content
    resp = model.generate_content(prompt)
    raw = resp.text or ""
    # 去掉可能的 ```json 包裹
    raw = raw.strip()
    if raw.startswith("```"):
        raw = raw.strip("`")
        raw = raw.split("\n", 1)[-1]
    if raw.endswith("```"):
        raw = raw[:-3]
    # 嘗試補齊 JSON
    try:
        return json.loads(raw)
    except json.JSONDecodeError:
        s = raw.find("{"); e = raw.rfind("}")
        if s != -1 and e != -1 and e > s:
            return json.loads(raw[s:e+1])
        raise

# 準備 LLM 輸入（只帶標題與 H2/H3）
context_blocks = []
for d in docs:
    ctx = {
        "url": d["url"],
        "title": d["title"],
        "h2": d["h2"],
        "h3": d["h3"]
    }
    context_blocks.append(ctx)

SYSTEM_PROMPT = """
あなたはSEOアナリストです。以下の入力は各ページの「タイトル」と「目次（H2/H3）」のみです。
各ページから、見出し中に**実際に現れるフレーズをそのまま**SEOサジェストとして抽出してください。

入力の各行は次の形式でタグ付きです:
[TITLE] ...       → source は "title"
[H2] ...          → source は "h2"
[H3] ...          → source は "h3"

厳守ルール:
- extraction_class は常に "seo_suggest"
- extraction_text は **原文一致の連続フレーズ**（2〜12文字）※生成・意訳禁止
- attributes には以下を含める:
  - source: "title" | "h2" | "h3"  （行頭タグから決める）
  - intent: "informational" | "navigational" | "commercial" | "transactional" のいずれか
  - facet: 代表カテゴリ（例: "アクセス/行き方","チケット/予約","料金/割引","営業時間/所要時間","注意点/ルール","体験/見どころ" 等）
- 抽出順は原文の出現順、重複は許容（後で集計するため）
- 見出しをまたいで抽出しない
"""

USER_PROMPT = textwrap.dedent(f"""
入力（ページごとに title/h2/h3 を与えます）:
{json.dumps(context_blocks, ensure_ascii=False, indent=2)}

期待する出力スキーマ:
{{
  "pages": [
    {{
      "url": "string",
      "suggestions": [
        {{
          "keyword": "string (原文一致)",
          "source": "title|h2|h3",
          "intent": "informational|navigational|commercial|transactional",
          "facet": "string"
        }}
      ]
    }}
  ]
}}
厳格に JSON のみを返してください。
""")

suggest_data = llm_json(GEMINI_ANALYSIS_MODEL, SYSTEM_PROMPT, USER_PROMPT, temperature=0.2)
suggest_data.keys(), len(suggest_data.get("pages", []))
````

````python
#Cell 5 — 統計：サジェスト出現次數（以「全文」統計），以及主查詢キーワード的出現次數
import pandas as pd

def count_occurrences(text: str, phrase: str) -> int:
    if not phrase or not isinstance(phrase, str):
        return 0
    return text.count(phrase)

# ---- A) サジェスト統計（全文 count）----
records = []
for pg in suggest_data.get("pages", []):
    url = pg.get("url")
    doc = next((d for d in docs if d["url"] == url), None)
    if not doc:
        continue
    full = doc["text"]
    for s in pg.get("suggestions", []):
        kw = s.get("keyword","").strip()
        if not kw: 
            continue
        records.append({
            "url": url,
            "domain": doc["domain"],
            "keyword": kw,
            "source": s.get("source",""),
            "intent": s.get("intent",""),
            "facet": s.get("facet",""),
            "occurrences_full": count_occurrences(full, kw)
        })

df_suggest = pd.DataFrame(records)
if df_suggest.empty:
    print("No suggestions extracted.")
else:
    # 聚合同 keyword across pages
    agg = (df_suggest.groupby(["keyword","intent","facet"], dropna=False)["occurrences_full"]
           .sum().reset_index().sort_values("occurrences_full", ascending=False))
    display(df_suggest.head(20))
    display(agg.head(30))

# ---- B) 主查詢キーワード的出現次數（每頁）----
q_records = []
for d in docs:
    for q in QUERY_KEYWORDS:
        q_records.append({
            "url": d["url"],
            "domain": d["domain"],
            "query": q,
            "occurrences_full": count_occurrences(d["text"], q)
        })
df_query = pd.DataFrame(q_records).sort_values(["query","occurrences_full"], ascending=[True, False])
display(df_query)
````

````python
# Cell — 將 Cell 4 & 5 的結果輸出成 Excel + HTML 一次搞定
import os, re, json, datetime
import pandas as pd
from pathlib import Path
from html import escape

# --------- 檔名工具 ---------
def slugify(s: str, max_len: int = 80) -> str:
    s = (s or "").strip()
    s = re.sub(r"[^\w\u4e00-\u9fff\u3040-\u30ff\u3400-\u4dbf\- ]+", "", s)  # 英數/中日/底線/空白/連字號
    s = re.sub(r"\s+", "_", s)
    return s[:max_len] or "seo_competitor_report"

ts = datetime.datetime.now().strftime("%Y%m%d_%H%M")
base = "seo_competitor_report"
xlsx_path = f"{base}_{ts}.xlsx"
html_path = f"{base}_{ts}.html"
json_path = f"{base}_{ts}.json"

# --------- 來源資料保護性存取 ---------
_docs = globals().get("docs", [])
_suggest_data = globals().get("suggest_data", {}) or {}
_df_suggest = globals().get("df_suggest", pd.DataFrame())
_df_query = globals().get("df_query", pd.DataFrame())
_QUERY_KEYWORDS = globals().get("QUERY_KEYWORDS", [])

# 轉成 DataFrame：頁面標題與目次（若要在 HTML 里展示）
def docs_to_df(docs_list):
    rows = []
    for d in docs_list:
        rows.append({
            "url": d.get("url",""),
            "domain": d.get("domain",""),
            "title": d.get("title",""),
            "h2": "\n".join(d.get("h2",[]) or []),
            "h3": "\n".join(d.get("h3",[]) or []),
            "text_len": len(d.get("text","") or "")
        })
    return pd.DataFrame(rows)

df_docs_overview = docs_to_df(_docs) if isinstance(_docs, list) and _docs else pd.DataFrame(
    columns=["url","domain","title","h2","h3","text_len"]
)

# 轉成 DataFrame：LLM 的建議（每頁）
def suggest_to_df(suggest_json):
    pages = suggest_json.get("pages", []) if isinstance(suggest_json, dict) else []
    rows = []
    for pg in pages:
        url = pg.get("url","")
        for s in pg.get("suggestions", []) or []:
            rows.append({
                "url": url,
                "keyword": s.get("keyword",""),
                "source": s.get("source",""),
                "intent": s.get("intent",""),
                "facet": s.get("facet","")
            })
    return pd.DataFrame(rows)

df_llm_suggest = suggest_to_df(_suggest_data)

# 聚合建議（Cell 5 的 agg 與 df_suggest 已經有；這裡保險做一份簡版聚合）
df_agg_simple = pd.DataFrame()
if not _df_suggest.empty:
    df_agg_simple = (
        _df_suggest.groupby(["keyword","intent","facet"], dropna=False)["occurrences_full"]
        .sum()
        .reset_index()
        .sort_values("occurrences_full", ascending=False)
    )

# --------- 輸出 Excel ---------
sheets = {}
sheets["docs_overview"] = df_docs_overview
if not df_llm_suggest.empty: sheets["llm_suggestions_title_h2h3"] = df_llm_suggest
if isinstance(_df_suggest, pd.DataFrame) and not _df_suggest.empty: sheets["suggest_full_counts"] = _df_suggest
if isinstance(df_agg_simple, pd.DataFrame) and not df_agg_simple.empty: sheets["suggest_full_counts_agg"] = df_agg_simple
if isinstance(_df_query, pd.DataFrame) and not _df_query.empty: sheets["query_counts_by_page"] = _df_query
if isinstance(_QUERY_KEYWORDS, list) and _QUERY_KEYWORDS: sheets["query_keywords"] = pd.DataFrame({"query": _QUERY_KEYWORDS})
# 保留原始 JSON
sheets["raw_suggest_json"] = pd.DataFrame({"json": [json.dumps(_suggest_data, ensure_ascii=False, indent=2)]})

with pd.ExcelWriter(xlsx_path, engine="openpyxl") as writer:
    for name, df in sheets.items():
        safe_name = re.sub(r"[^A-Za-z0-9_\-一-龥ぁ-んァ-ン]", "_", name)[:28] or "Sheet1"
        df.to_excel(writer, sheet_name=safe_name, index=False)

print(f"✅ 已輸出 Excel：{xlsx_path}")

# --------- 產生 HTML 報表 ---------
def render_html(docs_df, llm_sug_df, df_suggest_full, df_suggest_agg, df_query, query_keywords):
    style = """
    <style>
      body{font-family:-apple-system,BlinkMacSystemFont,"Segoe UI",Roboto,"Noto Sans JP","Hiragino Sans","Yu Gothic","Meiryo",Arial,sans-serif;line-height:1.6;padding:24px;color:#222;}
      h1{font-size:28px;margin:0 0 16px;}
      h2{font-size:22px;margin:24px 0 8px;border-left:4px solid #4F46E5;padding-left:8px;}
      .sec{margin-bottom:28px;}
      .card{background:#fff;border:1px solid #e5e7eb;border-radius:10px;padding:12px 14px;box-shadow:0 1px 2px rgba(0,0,0,0.04);}
      .muted{color:#6b7280;font-size:13px;}
      table{border-collapse:collapse;width:100%;font-size:14px;}
      th,td{border:1px solid #e5e7eb;padding:8px 10px;text-align:left;vertical-align:top;}
      th{background:#F9FAFB;}
      .tag{display:inline-block;background:#EEF2FF;color:#4338CA;padding:2px 8px;margin-right:6px;border-radius:999px;font-size:12px;}
      code{background:#F3F4F6;padding:2px 4px;border-radius:4px;}
    </style>
    """
    html = [f"<!doctype html><html lang='ja'><meta charset='utf-8'><title>SEO 競合サジェスト レポート</title>{style}<body>"]
    html.append("<h1>SEO 競合サジェスト レポート</h1>")
    html.append(f"<div class='muted'>Generated at {escape(ts)}</div>")

    # Docs overview
    html.append("<div class='sec'><h2>対象ページ（タイトル / 目次 概要）</h2>")
    if not docs_df.empty:
        html.append("<div class='card'><table><thead><tr><th>domain</th><th>url</th><th>title</th><th>H2（抜粋）</th><th>H3（抜粋）</th></tr></thead><tbody>")
        for _, r in docs_df.iterrows():
            h2_short = (r.get("h2") or "").split("\n")
            h3_short = (r.get("h3") or "").split("\n")
            h2_short = "<br>".join([escape(x) for x in h2_short[:6]])
            h3_short = "<br>".join([escape(x) for x in h3_short[:8]])
            html.append("<tr>"
                        f"<td>{escape(str(r.get('domain','')))}</td>"
                        f"<td><a href='{escape(str(r.get('url','')))}' target='_blank'>{escape(str(r.get('url','')))}</a></td>"
                        f"<td>{escape(str(r.get('title','')))}</td>"
                        f"<td>{h2_short}</td>"
                        f"<td>{h3_short}</td>"
                        "</tr>")
        html.append("</tbody></table></div>")
    else:
        html.append("<div class='card muted'>（データなし）</div>")
    html.append("</div>")

    # LLM suggestions from title/h2/h3
    html.append("<div class='sec'><h2>LLM抽出サジェスト（見出し由来）</h2>")
    if isinstance(llm_sug_df, pd.DataFrame) and not llm_sug_df.empty:
        show = llm_sug_df.copy().sort_values(["url","source","keyword"])
        html.append("<div class='card'><table><thead><tr><th>url</th><th>source</th><th>keyword</th><th>intent</th><th>facet</th></tr></thead><tbody>")
        for _, r in show.iterrows():
            html.append("<tr>"
                        f"<td><a href='{escape(str(r.get('url','')))}' target='_blank'>{escape(str(r.get('url','')))}</a></td>"
                        f"<td>{escape(str(r.get('source','')))}</td>"
                        f"<td><code>{escape(str(r.get('keyword','')))}</code></td>"
                        f"<td>{escape(str(r.get('intent','')))}</td>"
                        f"<td>{escape(str(r.get('facet','')))}</td>"
                        "</tr>")
        html.append("</tbody></table><div class='muted'>※ ここでは「見出しからの抽出結果」を表示。実際の全文出現回数は次セクションで。</div></div>")
    else:
        html.append("<div class='card muted'>（データなし）</div>")
    html.append("</div>")

    # Full-text counts for suggestions
    html.append("<div class='sec'><h2>サジェストの全文出現回数（集計）</h2>")
    if isinstance(df_suggest_agg, pd.DataFrame) and not df_suggest_agg.empty:
        show = df_suggest_agg.copy().sort_values("occurrences_full", ascending=False).head(200)
        html.append("<div class='card'><table><thead><tr><th>keyword</th><th>intent</th><th>facet</th><th>occurrences_full (合計)</th></tr></thead><tbody>")
        for _, r in show.iterrows():
            html.append("<tr>"
                        f"<td><code>{escape(str(r.get('keyword','')))}</code></td>"
                        f"<td>{escape(str(r.get('intent','')))}</td>"
                        f"<td>{escape(str(r.get('facet','')))}</td>"
                        f"<td>{int(r.get('occurrences_full') or 0)}</td>"
                        "</tr>")
        html.append("</tbody></table><div class='muted'>※ 上位200のみ表示。完全版は Excel で確認してください。</div></div>")
    else:
        html.append("<div class='card muted'>（データなし）</div>")
    html.append("</div>")

    # Per-page query counts
    html.append("<div class='sec'><h2>主クエリ（各ページ出現回数）</h2>")
    if isinstance(df_query, pd.DataFrame) and not df_query.empty:
        show = df_query.copy()
        show = show.sort_values(["query","occurrences_full"], ascending=[True, False]).head(300)
        html.append("<div class='card'><table><thead><tr><th>query</th><th>domain</th><th>url</th><th>occurrences_full</th></tr></thead><tbody>")
        for _, r in show.iterrows():
            html.append("<tr>"
                        f"<td><code>{escape(str(r.get('query','')))}</code></td>"
                        f"<td>{escape(str(r.get('domain','')))}</td>"
                        f"<td><a href='{escape(str(r.get('url','')))}' target='_blank'>{escape(str(r.get('url','')))}</a></td>"
                        f"<td>{int(r.get('occurrences_full') or 0)}</td>"
                        "</tr>")
        html.append("</tbody></table><div class='muted'>※ 上位300のみ表示。完全版は Excel で確認してください。</div></div>")
    else:
        html.append("<div class='card muted'>（データなし）</div>")
    html.append("</div>")

    # Query keywords list
    html.append("<div class='sec'><h2>主クエリ一覧</h2>")
    if isinstance(query_keywords, list) and query_keywords:
        html.append("<div class='card'><ul>")
        for q in query_keywords:
            html.append(f"<li><code>{escape(str(q))}</code></li>")
        html.append("</ul></div>")
    else:
        html.append("<div class='card muted'>（データなし）</div>")
    html.append("</div>")

    html.append("</body></html>")
    return "".join(html)

html_text = render_html(
    docs_df=df_docs_overview,
    llm_sug_df=df_llm_suggest,
    df_suggest_full=_df_suggest,
    df_suggest_agg=df_agg_simple,
    df_query=_df_query,
    query_keywords=_QUERY_KEYWORDS
)

with open(html_path, "w", encoding="utf-8") as f:
    f.write(html_text)

# 也保存所有原始資料（便於回放/比對）
with open(json_path, "w", encoding="utf-8") as f:
    json.dump({
        "docs": _docs,
        "suggest_data": _suggest_data,
        "df_suggest": _df_suggest.to_dict(orient="records") if isinstance(_df_suggest, pd.DataFrame) else None,
        "df_query": _df_query.to_dict(orient="records") if isinstance(_df_query, pd.DataFrame) else None,
        "query_keywords": _QUERY_KEYWORDS
    }, f, ensure_ascii=False, indent=2)

print(f"✅ 已輸出 HTML：{html_path}")
print(f"✅ 已輸出 JSON：{json_path}")
````

````python
# Cell — 將 Cell 6 的構成案 & 統計輸出成 Excel 與 HTML
import os, re, json, datetime
import pandas as pd
from pathlib import Path
from html import escape

# 檔名安全化
def slugify(s: str, max_len: int = 80) -> str:
    s = (s or "").strip()
    s = re.sub(r"[^\w\u4e00-\u9fff\u3040-\u30ff\u3400-\u4dbf\- ]+", "", s)  # 英數/中日文/底線/空白/連字號
    s = re.sub(r"\s+", "_", s)
    return s[:max_len] or "outline_plan"

# ===== 準備資料 =====
if 'outline_data' not in globals() or not isinstance(outline_data, dict):
    raise RuntimeError("找不到 outline_data。請先執行 Cell 6 產生構成案。")

# 可能來自 Cell 6 的 USER_BRIEF（若不存在就用空殼）
USER_BRIEF_SAFE = globals().get('USER_BRIEF', {})
TOP_SUGGESTS = USER_BRIEF_SAFE.get('top_suggests', [])
PER_PAGE_QUERY_COUNTS = USER_BRIEF_SAFE.get('per_page_query_counts', [])
QUERY_KEYWORDS = USER_BRIEF_SAFE.get('query_keywords', [])

# 產生檔名基底（用第一個 H2 當作 slug，否則預設）
first_h2 = None
if isinstance(outline_data.get("outline"), list) and outline_data["outline"]:
    first_h2 = outline_data["outline"][0].get("h2")
base = slugify(first_h2 or "outline_plan")
ts = datetime.datetime.now().strftime("%Y%m%d_%H%M")
xlsx_path = f"{base}_{ts}.xlsx"
html_path = f"{base}_{ts}.html"
json_path = f"{base}_{ts}.json"

# ===== 整理成 DataFrame =====
sheets = {}

# 1) outline → DataFrame（H2/H3展開；example_keywords 合併顯示）
def outline_to_rows(outline_list):
    rows = []
    if not isinstance(outline_list, list):
        return rows
    for blk in outline_list:
        if not isinstance(blk, dict):
            continue
        h2 = blk.get("h2") or ""
        why = blk.get("why") or ""
        intent = blk.get("intent") or ""
        facet = blk.get("facet") or ""
        h3s = blk.get("h3") or []
        exkws = blk.get("example_keywords") or []
        # 若沒有 h3，至少也輸出 H2 層級
        if not h3s:
            rows.append({
                "H2": h2,
                "H3": "",
                "intent": intent,
                "facet": facet,
                "why": why,
                "example_keywords": "、".join([str(k).strip() for k in exkws if str(k).strip()]),
            })
        else:
            for h3 in h3s:
                rows.append({
                    "H2": h2,
                    "H3": h3,
                    "intent": intent,
                    "facet": facet,
                    "why": why,
                    "example_keywords": "、".join([str(k).strip() for k in exkws if str(k).strip()]),
                })
    return rows

outline_rows = outline_to_rows(outline_data.get("outline", []))
df_outline = pd.DataFrame(outline_rows) if outline_rows else pd.DataFrame(columns=["H2","H3","intent","facet","why","example_keywords"])
sheets["outline"] = df_outline

# 2) FAQs
faqs = outline_data.get("faqs", [])
if isinstance(faqs, list) and faqs:
    df_faqs = pd.DataFrame({"faq": [str(x) for x in faqs]})
    sheets["faqs"] = df_faqs

# 3) CTA
cta = outline_data.get("cta")
if cta:
    sheets["cta"] = pd.DataFrame({"cta": [str(cta)]})

# 4) 內部連結
internal_linking = outline_data.get("internal_linking", [])
if isinstance(internal_linking, list) and internal_linking:
    sheets["internal_links"] = pd.DataFrame({"internal_link": [str(x) for x in internal_linking]})

# 5) 編輯注意
notes = outline_data.get("notes", [])
if isinstance(notes, list) and notes:
    sheets["notes"] = pd.DataFrame({"note": [str(x) for x in notes]})

# 6) top_suggests（來自 USER_BRIEF）
if isinstance(TOP_SUGGESTS, list) and TOP_SUGGESTS:
    # 欄位：keyword, intent, facet, occurrences_full
    df_top_suggests = pd.DataFrame(TOP_SUGGESTS)
    sheets["top_suggests"] = df_top_suggests

# 7) 各頁主查詢關鍵字出現次數（per_page_query_counts）
if isinstance(PER_PAGE_QUERY_COUNTS, list) and PER_PAGE_QUERY_COUNTS:
    df_page_counts = pd.DataFrame(PER_PAGE_QUERY_COUNTS)
    sheets["per_page_query_counts"] = df_page_counts

# 8) 主查詢關鍵字清單（query_keywords）
if isinstance(QUERY_KEYWORDS, list) and QUERY_KEYWORDS:
    df_qk = pd.DataFrame({"query_keyword": [str(x) for x in QUERY_KEYWORDS]})
    sheets["query_keywords"] = df_qk

# 9) 保留原始 JSON
sheets["raw_outline_json"] = pd.DataFrame({"json": [json.dumps(outline_data, ensure_ascii=False, indent=2)]})
sheets["raw_user_brief"] = pd.DataFrame({"json": [json.dumps(USER_BRIEF_SAFE, ensure_ascii=False, indent=2)]})

# ===== 輸出 Excel =====
with pd.ExcelWriter(xlsx_path, engine="openpyxl") as writer:
    for name, df in sheets.items():
        safe_name = re.sub(r"[^A-Za-z0-9_\-一-龥ぁ-んァ-ン]", "_", name)[:28] or "Sheet1"
        df.to_excel(writer, sheet_name=safe_name, index=False)

print(f"✅ 已輸出 Excel：{xlsx_path}")

# ===== 產生可視化 HTML =====
def render_html(outline_df, faqs_list, cta_text, ilinks_list, notes_list, top_sug_df, page_counts_df, qk_list):
    # 基本樣式
    style = """
    <style>
      body{font-family: -apple-system, BlinkMacSystemFont, "Segoe UI", Roboto, "Noto Sans JP","Hiragino Sans","Yu Gothic", "Meiryo", Arial, sans-serif; line-height:1.6; padding:24px; color:#222;}
      h1{font-size:28px; margin:0 0 16px;}
      h2{font-size:22px; margin:24px 0 8px; border-left:4px solid #4F46E5; padding-left:8px;}
      h3{font-size:18px; margin:12px 0 4px;}
      .sec{margin-bottom:28px;}
      .card{background:#fff; border:1px solid #e5e7eb; border-radius:10px; padding:12px 14px; box-shadow: 0 1px 2px rgba(0,0,0,0.04);}
      .tag{display:inline-block; background:#EEF2FF; color:#4338CA; padding:2px 8px; margin-right:6px; border-radius:999px; font-size:12px;}
      table{border-collapse:collapse; width:100%; font-size:14px;}
      th,td{border:1px solid #e5e7eb; padding:8px 10px; text-align:left; vertical-align:top;}
      th{background:#F9FAFB;}
      .muted{color:#6b7280; font-size:13px;}
      ul{margin:6px 0 12px 20px;}
      code{background:#F3F4F6; padding:2px 4px; border-radius:4px;}
    </style>
    """
    html = [f"<!doctype html><html lang='ja'><meta charset='utf-8'><title>{escape(first_h2 or '記事構成案プレビュー')}</title>{style}<body>"]
    html.append(f"<h1>記事構成案プレビュー</h1>")
    html.append("<div class='sec card'><div class='muted'>このページは Cell 6 の出力（構成案・統計）を見やすくまとめたプレビューです。</div></div>")

    # Outline（H2/H3）
    html.append("<div class='sec'><h2>H2/H3 構成</h2>")
    if not outline_df.empty:
        current_h2 = None
        for _, row in outline_df.iterrows():
            h2 = str(row.get("H2") or "")
            h3 = str(row.get("H3") or "")
            intent = str(row.get("intent") or "")
            facet = str(row.get("facet") or "")
            why = str(row.get("why") or "")
            exkws = str(row.get("example_keywords") or "")

            if h2 != current_h2:
                if current_h2 is not None:
                    html.append("</div>")  # close previous h2 block
                html.append(f"<div class='card' style='margin-bottom:14px'><h3>{escape(h2)}</h3>")
                if intent or facet:
                    html.append("<div>")
                    if intent: html.append(f"<span class='tag'>intent: {escape(intent)}</span>")
                    if facet:  html.append(f"<span class='tag'>facet: {escape(facet)}</span>")
                    html.append("</div>")
                if why:
                    html.append(f"<div class='muted' style='margin-top:6px'>{escape(why)}</div>")
                if exkws:
                    html.append(f"<div class='muted' style='margin-top:6px'><b>例キーワード:</b> {escape(exkws)}</div>")
                current_h2 = h2

            if h3:
                html.append(f"<div style='margin-left:12px; padding-left:10px; border-left:2px solid #E5E7EB; margin-top:6px'>{escape(h3)}</div>")
        if current_h2 is not None:
            html.append("</div>")
    else:
        html.append("<div class='card'>（構成案が空です）</div>")
    html.append("</div>")

    # FAQs
    html.append("<div class='sec'><h2>FAQ</h2>")
    if isinstance(faqs_list, list) and faqs_list:
        html.append("<div class='card'><ul>")
        for q in faqs_list:
            html.append(f"<li>{escape(str(q))}</li>")
        html.append("</ul></div>")
    else:
        html.append("<div class='card muted'>なし</div>")
    html.append("</div>")

    # CTA
    html.append("<div class='sec'><h2>CTA</h2>")
    if cta_text:
        html.append(f"<div class='card'>{escape(str(cta_text))}</div>")
    else:
        html.append("<div class='card muted'>なし</div>")
    html.append("</div>")

    # 內部連結
    html.append("<div class='sec'><h2>内部リンク案</h2>")
    if isinstance(ilinks_list, list) and ilinks_list:
        html.append("<div class='card'><ul>")
        for it in ilinks_list:
            html.append(f"<li>{escape(str(it))}</li>")
        html.append("</ul></div>")
    else:
        html.append("<div class='card muted'>なし</div>")
    html.append("</div>")

    # Top suggests
    html.append("<div class='sec'><h2>競合サジェスト（上位・全文出現回数）</h2>")
    if isinstance(top_sug_df, pd.DataFrame) and not top_sug_df.empty:
        # 排序顯示
        ts_view = top_sug_df.copy().sort_values("occurrences_full", ascending=False)
        html.append("<div class='card'><table><thead><tr><th>keyword</th><th>intent</th><th>facet</th><th>occurrences_full</th></tr></thead><tbody>")
        for _, r in ts_view.iterrows():
            html.append(f"<tr><td>{escape(str(r.get('keyword','')))}</td>"
                        f"<td>{escape(str(r.get('intent','')))}</td>"
                        f"<td>{escape(str(r.get('facet','')))}</td>"
                        f"<td>{int(r.get('occurrences_full') or 0)}</td></tr>")
        html.append("</tbody></table></div>")
    else:
        html.append("<div class='card muted'>なし</div>")
    html.append("</div>")

    # Per-page query counts
    html.append("<div class='sec'><h2>各ページの主クエリ出現回数</h2>")
    if isinstance(page_counts_df, pd.DataFrame) and not page_counts_df.empty:
        # 只顯示前幾欄，避免太寬
        cols = list(page_counts_df.columns)
        keep = cols[: min(8, len(cols))]
        show_df = page_counts_df[keep].copy()
        html.append("<div class='card'><table><thead><tr>")
        for c in show_df.columns:
            html.append(f"<th>{escape(str(c))}</th>")
        html.append("</tr></thead><tbody>")
        for _, r in show_df.iterrows():
            html.append("<tr>" + "".join(f"<td>{escape(str(v))}</td>" for v in r.values) + "</tr>")
        html.append("</tbody></table><div class='muted'>※ 表示は前 8 欄のみ。Excel で完全版をご確認ください。</div></div>")
    else:
        html.append("<div class='card muted'>なし</div>")
    html.append("</div>")

    # Query keywords
    html.append("<div class='sec'><h2>主クエリ一覧</h2>")
    if isinstance(qk_list, list) and qk_list:
        html.append("<div class='card'><ul>")
        for k in qk_list:
            html.append(f"<li><code>{escape(str(k))}</code></li>")
        html.append("</ul></div>")
    else:
        html.append("<div class='card muted'>なし</div>")
    html.append("</div>")

    html.append("<div class='sec muted'>Generated at " + escape(ts) + "</div>")
    html.append("</body></html>")
    return "".join(html)

html_text = render_html(
    outline_df=df_outline,
    faqs_list=faqs,
    cta_text=cta,
    ilinks_list=internal_linking,
    notes_list=notes,
    top_sug_df=sheets.get("top_suggests"),
    page_counts_df=sheets.get("per_page_query_counts"),
    qk_list=QUERY_KEYWORDS,
)

with open(html_path, "w", encoding="utf-8") as f:
    f.write(html_text)

# 也一起保存原始 JSON，方便溯源
with open(json_path, "w", encoding="utf-8") as f:
    json.dump({"outline_data": outline_data, "user_brief": USER_BRIEF_SAFE}, f, ensure_ascii=False, indent=2)

print(f"✅ 已輸出 HTML：{html_path}")
print(f"✅ 已輸出 JSON：{json_path}")
````

````python
#Cell 4 — LangExtract：從「標題 + H2/H3」抽取 競品が使っているサジェスト（不傳全文）
# Cell 4 (LangExtract版) —— 只用 Title/H2/H3 來抽取「exact サジェスト」
# 需求：docs = [{"url","domain","title","h2":[...],"h3":[...],"text":全文}, ...] 已就緒

import langextract as lx
import textwrap, json, os

# 你可用先前 notebook 的設定，或在這裡覆寫
LX_MODEL_ID = globals().get("GEMINI_ANALYSIS_MODEL", "gemini-2.5-flash")

# —— 提示詞（強制 exact 抽取 + 來源標註）
lx_prompt = textwrap.dedent("""
あなたはSEOアナリストです。以下の入力は各ページの「タイトル」と「目次（H2/H3）」のみです。
各ページから、見出し中に**実際に現れるフレーズをそのまま**SEOサジェストとして抽出してください。

入力の各行は次の形式でタグ付きです:
[TITLE] ...       → source は "title"
[H2] ...          → source は "h2"
[H3] ...          → source は "h3"

厳守ルール:
- extraction_class は常に "seo_suggest"
- extraction_text は **原文一致の連続フレーズ**（2〜12文字）※生成・意訳禁止
- attributes には以下を含める:
  - source: "title" | "h2" | "h3"  （行頭タグから決める）
  - intent: "informational" | "navigational" | "commercial" | "transactional" のいずれか
  - facet: 代表カテゴリ（例: "アクセス/行き方","チケット/予約","料金/割引","営業時間/所要時間","注意点/ルール","体験/見どころ" 等）
- 抽出順は原文の出現順、重複は許容（後で集計するため）
- 見出しをまたいで抽出しない
""").strip()

# —— Few-shot：用極小片段教會「exact + source/intent/facet」
example_text = "[TITLE] 東京ハリー・ポッター スタジオツアー｜チケットの買い方とアクセス\n[H2] チケットの買い方\n[H3] 当日券はある？"
lx_examples = [
    lx.data.ExampleData(
        text=example_text,
        extractions=[
            lx.data.Extraction(
                extraction_class="seo_suggest",
                extraction_text="チケットの買い方",  # ← タイトル内の exact 文字列
                attributes={"source":"title","intent":"transactional","facet":"チケット/予約"}
            ),
            lx.data.Extraction(
                extraction_class="seo_suggest",
                extraction_text="アクセス",           # ← タイトル内の exact 文字列
                attributes={"source":"title","intent":"informational","facet":"アクセス/行き方"}
            ),
            lx.data.Extraction(
                extraction_class="seo_suggest",
                extraction_text="当日券",            # ← H3 内の exact 文字列
                attributes={"source":"h3","intent":"transactional","facet":"チケット/予約"}
            ),
        ]
    )
]

def build_tagged_outline(d):
    lines = []
    if d.get("title"): lines.append(f"[TITLE] {d['title']}")
    for h in d.get("h2", []) or []:
        if h: lines.append(f"[H2] {h}")
    for h in d.get("h3", []) or []:
        if h: lines.append(f"[H3] {h}")
    return "\n".join(lines)

# —— 逐頁抽取（只送出 title/H2/H3）
pages = []
for d in docs:
    tagged = build_tagged_outline(d)
    if not tagged.strip():
        pages.append({"url": d["url"], "suggestions": []})
        continue

    # LangExtract：用 Gemini 進行抽取（多pass 提升 recall；內容很短，buffer 可小）
    import os

    API_KEY = os.getenv("LANGEXTRACT_API_KEY") or os.getenv("GOOGLE_API_KEY")
    if not API_KEY:
        raise RuntimeError("No API key. Set LANGEXTRACT_API_KEY or GOOGLE_API_KEY")
    
    res = lx.extract(
        text_or_documents=tagged,
        prompt_description=lx_prompt,
        examples=lx_examples,
        model_id=LX_MODEL_ID,          # 例如 "gemini-2.5-flash"
        api_key=API_KEY,               # ←← 關鍵：直接傳給 LangExtract
        extraction_passes=2,
        max_workers=4,
        max_char_buffer=600
    )


    sug_list = []
    for e in res.extractions:
        if e.extraction_class != "seo_suggest":
            continue
        kw = (e.extraction_text or "").strip()
        if not kw:
            continue
        attrs = e.attributes or {}
        src = (attrs.get("source") or "").lower()
        if src not in ("title","h2","h3"):
            # 万一 source を付け忘れたら、見出しタグで類推しない（厳格運用）→ "h2" を仮置き
            src = "h2"
        sug_list.append({
            "keyword": kw,
            "source": src,
            "intent": attrs.get("intent"),
            "facet": attrs.get("facet"),
        })
    pages.append({"url": d["url"], "suggestions": sug_list})

# 與你原本變數名對齊：suggest_data
suggest_data = {"pages": pages}
print("LangExtract pages:", len(suggest_data["pages"]))
# 可選：print 一小段檢查
if suggest_data["pages"]:
    print(json.dumps(suggest_data["pages"][0], ensure_ascii=False, indent=2)[:600], "...")
````

````python
#Cell 5 — 統計：サジェストの全文出現回数 & 主クエリ出現回数（沿用你原本的統計寫法）

# Cell 5 —— 統計（沿用你的欄位；df_suggest/agg/df_query 會被後續流程使用）
import pandas as pd

def count_occurrences(text: str, phrase: str) -> int:
    if not phrase or not isinstance(phrase, str):
        return 0
    return text.count(phrase)

# ---- A) サジェスト統計（全文 count）----
records = []
for pg in suggest_data.get("pages", []):
    url = pg.get("url")
    doc = next((d for d in docs if d["url"] == url), None)
    if not doc:
        continue
    full = doc.get("text","") or ""
    for s in pg.get("suggestions", []) or []:
        kw = (s.get("keyword") or "").strip()
        if not kw:
            continue
        records.append({
            "url": url,
            "domain": doc.get("domain",""),
            "keyword": kw,
            "source": s.get("source",""),
            "intent": s.get("intent",""),
            "facet": s.get("facet",""),
            "occurrences_full": count_occurrences(full, kw)
        })

df_suggest = pd.DataFrame(records)
if df_suggest.empty:
    print("No suggestions extracted.")
else:
    agg = (
        df_suggest.groupby(["keyword","intent","facet"], dropna=False)["occurrences_full"]
        .sum()
        .reset_index()
        .sort_values("occurrences_full", ascending=False)
    )
    display(df_suggest.head(20))
    display(agg.head(30))

# ---- B) 主クエリ（每頁出現次數）----
q_records = []
for d in docs:
    full = d.get("text","") or ""
    for q in QUERY_KEYWORDS:
        q_records.append({
            "url": d["url"],
            "domain": d["domain"],
            "query": q,
            "occurrences_full": count_occurrences(full, q)
        })
df_query = pd.DataFrame(q_records).sort_values(["query","occurrences_full"], ascending=[True, False])
display(df_query)
````

````python
# === Cell 6 — 產生「記事構成案」（LLM：根據主查詢 + サジェスト + 統計） ===
import os, json, textwrap
import pandas as pd

# ---------- 安全檢查 ----------
API_KEY = os.getenv("LANGEXTRACT_API_KEY") or os.getenv("GOOGLE_API_KEY")
if not API_KEY:
    raise RuntimeError("No API key found. Please set GOOGLE_API_KEY or LANGEXTRACT_API_KEY.")

# 基礎變數保底
QUERY_KEYWORDS = globals().get("QUERY_KEYWORDS", []) or []
df_suggest = df_suggest.copy() if 'df_suggest' in globals() else pd.DataFrame(columns=["keyword","intent","facet","occurrences_full"])
df_query   = df_query.copy()   if 'df_query'   in globals() else pd.DataFrame(columns=["url","domain","query","occurrences_full"])

# ---------- 壓縮：取 Top N 的競合サジェスト ----------
TOP_N_SUGGEST = 60  # 你可調整
agg_compact = []
if not df_suggest.empty:
    # 聚合同一 keyword 在不同頁的出現次數
    top_sug = (
        df_suggest
        .groupby(["keyword","intent","facet"], dropna=False)["occurrences_full"]
        .sum()
        .reset_index()
        .sort_values("occurrences_full", ascending=False)
        .head(TOP_N_SUGGEST)
    )
    for _, r in top_sug.iterrows():
        intent = r["intent"] if pd.notna(r["intent"]) else "informational"
        facet  = r["facet"]  if pd.notna(r["facet"])  else ""
        agg_compact.append({
            "keyword": str(r["keyword"]),
            "intent":  str(intent),
            "facet":   str(facet),
            "occurrences_full": int(r["occurrences_full"] if pd.notna(r["occurrences_full"]) else 0)
        })

# ---------- 壓縮：各頁主查詢關鍵字出現次數（保持原表即可） ----------
per_page_query_counts = df_query.to_dict(orient="records") if not df_query.empty else []

# ---------- 整備 llm_json（若先前未定義則提供一個） ----------
try:
    _ = llm_json  # 檢查是否已存在
except NameError:
    import google.generativeai as genai
    genai.configure(api_key=API_KEY)

    def llm_json(model_name: str, system_prompt: str, user_content: str, temperature=0.3):
        """以 application/json 要求模型回傳 JSON；同時容錯去掉 ```json 區塊。"""
        model = genai.GenerativeModel(
            model_name,
            generation_config={
                "temperature": temperature,
                "response_mime_type": "application/json",
            }
        )
        prompt = system_prompt + "\n\n" + user_content
        resp = model.generate_content(prompt)
        raw = (resp.text or "").strip()

        # 去掉可能的三引號包裹
        if raw.startswith("```"):
            raw = raw.strip("`")
            raw = raw.split("\n", 1)[-1]
        if raw.endswith("```"):
            raw = raw[:-3]

        # 解析 JSON（容錯擷取最外層大括號）
        try:
            return json.loads(raw)
        except json.JSONDecodeError:
            s = raw.find("{"); e = raw.rfind("}")
            if s != -1 and e != -1 and e > s:
                return json.loads(raw[s:e+1])
            raise

# ---------- 準備 USER_BRIEF，盡量輕量以避免超額 tokens ----------
USER_BRIEF = {
    "query_keywords": QUERY_KEYWORDS,
    "top_suggests": agg_compact,                  # 已聚合 + 取前 N
    "per_page_query_counts": per_page_query_counts
}

# ---------- 系統提示（要求產出 H2/H3 構成案 + FAQ/CTA/内部リンク/注意；只回 JSON） ----------
SYS2 = textwrap.dedent("""\
    あなたは上級SEOストラテジストです。以下の「主クエリ」「競合のサジェスト（全文出現回数付き）」
    「各ページの主クエリ出現回数」に基づき、日本語で**実践的な記事構成案（H2/H3レベル）**を提案してください。

    要件:
    - ユーザーの検索意図（informational / navigational / commercial / transactional）を自然にカバー
    - 各H2に「why（なぜこの見出しが必要か）」を添える
    - 各H2に関連する例キーワード（example_keywords）を数個提示
    - 内部リンク案（サイト内の想定カテゴリ/関連記事の方向性）も列挙
    - FAQを5～10個、CTAを1段落、編集上の注意（E-E-A-T、更新頻度/価格変動の扱い等）を列挙

    出力は JSON のみ。以下スキーマに厳格準拠:
    {
      "outline": [
        {"h2": "string", "why": "string", "intent": "string", "facet": "string", "h3": ["string", "..."], "example_keywords": ["string","..."]}
      ],
      "faqs": ["string","..."],
      "cta": "string",
      "internal_linking": ["string","..."],
      "notes": ["string","..."]
    }
""")

# ---------- 使用者內容（把精簡後的資料丟入；請模型只回 JSON） ----------
USR2 = "入力:\n" + json.dumps(USER_BRIEF, ensure_ascii=False, indent=2) + "\n厳格に JSON のみを返してください。"

# ---------- 選擇模型 ----------
GEMINI_ANALYSIS_MODEL = globals().get("GEMINI_ANALYSIS_MODEL", "gemini-2.5-flash")

# ---------- 呼叫 LLM 產出構成案 ----------
outline_data = llm_json(GEMINI_ANALYSIS_MODEL, SYS2, USR2, temperature=0.4)

# ---------- 保底鍵值 ----------
if not isinstance(outline_data, dict):
    raise RuntimeError("LLM did not return a valid JSON object.")
outline_data.setdefault("outline", [])
outline_data.setdefault("faqs", [])
outline_data.setdefault("cta", "")
outline_data.setdefault("internal_linking", [])
outline_data.setdefault("notes", [])

print("✅ 構成案產生完成")
print(" - H2/H3 ブロック:", len(outline_data.get("outline", [])))
print(" - FAQ 件数:", len(outline_data.get("faqs", [])))
print(" - 内部リンク案:", len(outline_data.get("internal_linking", [])))
print(" - 注意事項:", len(outline_data.get("notes", [])))
````

````python
# Cell — 將 Cell 6 的構成案 & 統計輸出成 Excel 與 HTML
import os, re, json, datetime
import pandas as pd
from pathlib import Path
from html import escape

# 檔名安全化
def slugify(s: str, max_len: int = 80) -> str:
    s = (s or "").strip()
    s = re.sub(r"[^\w\u4e00-\u9fff\u3040-\u30ff\u3400-\u4dbf\- ]+", "", s)  # 英數/中日文/底線/空白/連字號
    s = re.sub(r"\s+", "_", s)
    return s[:max_len] or "outline_plan"

# ===== 準備資料 =====
if 'outline_data' not in globals() or not isinstance(outline_data, dict):
    raise RuntimeError("找不到 outline_data。請先執行 Cell 6 產生構成案。")

# 可能來自 Cell 6 的 USER_BRIEF（若不存在就用空殼）
USER_BRIEF_SAFE = globals().get('USER_BRIEF', {})
TOP_SUGGESTS = USER_BRIEF_SAFE.get('top_suggests', [])
PER_PAGE_QUERY_COUNTS = USER_BRIEF_SAFE.get('per_page_query_counts', [])
QUERY_KEYWORDS = USER_BRIEF_SAFE.get('query_keywords', [])

# 產生檔名基底（用第一個 H2 當作 slug，否則預設）
first_h2 = None
if isinstance(outline_data.get("outline"), list) and outline_data["outline"]:
    first_h2 = outline_data["outline"][0].get("h2")
base = slugify(first_h2 or "outline_plan")
ts = datetime.datetime.now().strftime("%Y%m%d_%H%M")
xlsx_path = f"LangExtract{base}_{ts}.xlsx"
html_path = f"LangExtract{base}_{ts}.html"
json_path = f"LangExtract{base}_{ts}.json"

# ===== 整理成 DataFrame =====
sheets = {}

# 1) outline → DataFrame（H2/H3展開；example_keywords 合併顯示）
def outline_to_rows(outline_list):
    rows = []
    if not isinstance(outline_list, list):
        return rows
    for blk in outline_list:
        if not isinstance(blk, dict):
            continue
        h2 = blk.get("h2") or ""
        why = blk.get("why") or ""
        intent = blk.get("intent") or ""
        facet = blk.get("facet") or ""
        h3s = blk.get("h3") or []
        exkws = blk.get("example_keywords") or []
        # 若沒有 h3，至少也輸出 H2 層級
        if not h3s:
            rows.append({
                "H2": h2,
                "H3": "",
                "intent": intent,
                "facet": facet,
                "why": why,
                "example_keywords": "、".join([str(k).strip() for k in exkws if str(k).strip()]),
            })
        else:
            for h3 in h3s:
                rows.append({
                    "H2": h2,
                    "H3": h3,
                    "intent": intent,
                    "facet": facet,
                    "why": why,
                    "example_keywords": "、".join([str(k).strip() for k in exkws if str(k).strip()]),
                })
    return rows

outline_rows = outline_to_rows(outline_data.get("outline", []))
df_outline = pd.DataFrame(outline_rows) if outline_rows else pd.DataFrame(columns=["H2","H3","intent","facet","why","example_keywords"])
sheets["outline"] = df_outline

# 2) FAQs
faqs = outline_data.get("faqs", [])
if isinstance(faqs, list) and faqs:
    df_faqs = pd.DataFrame({"faq": [str(x) for x in faqs]})
    sheets["faqs"] = df_faqs

# 3) CTA
cta = outline_data.get("cta")
if cta:
    sheets["cta"] = pd.DataFrame({"cta": [str(cta)]})

# 4) 內部連結
internal_linking = outline_data.get("internal_linking", [])
if isinstance(internal_linking, list) and internal_linking:
    sheets["internal_links"] = pd.DataFrame({"internal_link": [str(x) for x in internal_linking]})

# 5) 編輯注意
notes = outline_data.get("notes", [])
if isinstance(notes, list) and notes:
    sheets["notes"] = pd.DataFrame({"note": [str(x) for x in notes]})

# 6) top_suggests（來自 USER_BRIEF）
if isinstance(TOP_SUGGESTS, list) and TOP_SUGGESTS:
    # 欄位：keyword, intent, facet, occurrences_full
    df_top_suggests = pd.DataFrame(TOP_SUGGESTS)
    sheets["top_suggests"] = df_top_suggests

# 7) 各頁主查詢關鍵字出現次數（per_page_query_counts）
if isinstance(PER_PAGE_QUERY_COUNTS, list) and PER_PAGE_QUERY_COUNTS:
    df_page_counts = pd.DataFrame(PER_PAGE_QUERY_COUNTS)
    sheets["per_page_query_counts"] = df_page_counts

# 8) 主查詢關鍵字清單（query_keywords）
if isinstance(QUERY_KEYWORDS, list) and QUERY_KEYWORDS:
    df_qk = pd.DataFrame({"query_keyword": [str(x) for x in QUERY_KEYWORDS]})
    sheets["query_keywords"] = df_qk

# 9) 保留原始 JSON
sheets["raw_outline_json"] = pd.DataFrame({"json": [json.dumps(outline_data, ensure_ascii=False, indent=2)]})
sheets["raw_user_brief"] = pd.DataFrame({"json": [json.dumps(USER_BRIEF_SAFE, ensure_ascii=False, indent=2)]})

# ===== 輸出 Excel =====
with pd.ExcelWriter(xlsx_path, engine="openpyxl") as writer:
    for name, df in sheets.items():
        safe_name = re.sub(r"[^A-Za-z0-9_\-一-龥ぁ-んァ-ン]", "_", name)[:28] or "Sheet1"
        df.to_excel(writer, sheet_name=safe_name, index=False)

print(f"✅ 已輸出 Excel：{xlsx_path}")

# ===== 產生可視化 HTML =====
def render_html(outline_df, faqs_list, cta_text, ilinks_list, notes_list, top_sug_df, page_counts_df, qk_list):
    # 基本樣式
    style = """
    <style>
      body{font-family: -apple-system, BlinkMacSystemFont, "Segoe UI", Roboto, "Noto Sans JP","Hiragino Sans","Yu Gothic", "Meiryo", Arial, sans-serif; line-height:1.6; padding:24px; color:#222;}
      h1{font-size:28px; margin:0 0 16px;}
      h2{font-size:22px; margin:24px 0 8px; border-left:4px solid #4F46E5; padding-left:8px;}
      h3{font-size:18px; margin:12px 0 4px;}
      .sec{margin-bottom:28px;}
      .card{background:#fff; border:1px solid #e5e7eb; border-radius:10px; padding:12px 14px; box-shadow: 0 1px 2px rgba(0,0,0,0.04);}
      .tag{display:inline-block; background:#EEF2FF; color:#4338CA; padding:2px 8px; margin-right:6px; border-radius:999px; font-size:12px;}
      table{border-collapse:collapse; width:100%; font-size:14px;}
      th,td{border:1px solid #e5e7eb; padding:8px 10px; text-align:left; vertical-align:top;}
      th{background:#F9FAFB;}
      .muted{color:#6b7280; font-size:13px;}
      ul{margin:6px 0 12px 20px;}
      code{background:#F3F4F6; padding:2px 4px; border-radius:4px;}
    </style>
    """
    html = [f"<!doctype html><html lang='ja'><meta charset='utf-8'><title>{escape(first_h2 or '記事構成案プレビュー')}</title>{style}<body>"]
    html.append(f"<h1>記事構成案プレビュー</h1>")
    html.append("<div class='sec card'><div class='muted'>このページは Cell 6 の出力（構成案・統計）を見やすくまとめたプレビューです。</div></div>")

    # Outline（H2/H3）
    html.append("<div class='sec'><h2>H2/H3 構成</h2>")
    if not outline_df.empty:
        current_h2 = None
        for _, row in outline_df.iterrows():
            h2 = str(row.get("H2") or "")
            h3 = str(row.get("H3") or "")
            intent = str(row.get("intent") or "")
            facet = str(row.get("facet") or "")
            why = str(row.get("why") or "")
            exkws = str(row.get("example_keywords") or "")

            if h2 != current_h2:
                if current_h2 is not None:
                    html.append("</div>")  # close previous h2 block
                html.append(f"<div class='card' style='margin-bottom:14px'><h3>{escape(h2)}</h3>")
                if intent or facet:
                    html.append("<div>")
                    if intent: html.append(f"<span class='tag'>intent: {escape(intent)}</span>")
                    if facet:  html.append(f"<span class='tag'>facet: {escape(facet)}</span>")
                    html.append("</div>")
                if why:
                    html.append(f"<div class='muted' style='margin-top:6px'>{escape(why)}</div>")
                if exkws:
                    html.append(f"<div class='muted' style='margin-top:6px'><b>例キーワード:</b> {escape(exkws)}</div>")
                current_h2 = h2

            if h3:
                html.append(f"<div style='margin-left:12px; padding-left:10px; border-left:2px solid #E5E7EB; margin-top:6px'>{escape(h3)}</div>")
        if current_h2 is not None:
            html.append("</div>")
    else:
        html.append("<div class='card'>（構成案が空です）</div>")
    html.append("</div>")

    # FAQs
    html.append("<div class='sec'><h2>FAQ</h2>")
    if isinstance(faqs_list, list) and faqs_list:
        html.append("<div class='card'><ul>")
        for q in faqs_list:
            html.append(f"<li>{escape(str(q))}</li>")
        html.append("</ul></div>")
    else:
        html.append("<div class='card muted'>なし</div>")
    html.append("</div>")

    # CTA
    html.append("<div class='sec'><h2>CTA</h2>")
    if cta_text:
        html.append(f"<div class='card'>{escape(str(cta_text))}</div>")
    else:
        html.append("<div class='card muted'>なし</div>")
    html.append("</div>")

    # 內部連結
    html.append("<div class='sec'><h2>内部リンク案</h2>")
    if isinstance(ilinks_list, list) and ilinks_list:
        html.append("<div class='card'><ul>")
        for it in ilinks_list:
            html.append(f"<li>{escape(str(it))}</li>")
        html.append("</ul></div>")
    else:
        html.append("<div class='card muted'>なし</div>")
    html.append("</div>")

    # Top suggests
    html.append("<div class='sec'><h2>競合サジェスト（上位・全文出現回数）</h2>")
    if isinstance(top_sug_df, pd.DataFrame) and not top_sug_df.empty:
        # 排序顯示
        ts_view = top_sug_df.copy().sort_values("occurrences_full", ascending=False)
        html.append("<div class='card'><table><thead><tr><th>keyword</th><th>intent</th><th>facet</th><th>occurrences_full</th></tr></thead><tbody>")
        for _, r in ts_view.iterrows():
            html.append(f"<tr><td>{escape(str(r.get('keyword','')))}</td>"
                        f"<td>{escape(str(r.get('intent','')))}</td>"
                        f"<td>{escape(str(r.get('facet','')))}</td>"
                        f"<td>{int(r.get('occurrences_full') or 0)}</td></tr>")
        html.append("</tbody></table></div>")
    else:
        html.append("<div class='card muted'>なし</div>")
    html.append("</div>")

    # Per-page query counts
    html.append("<div class='sec'><h2>各ページの主クエリ出現回数</h2>")
    if isinstance(page_counts_df, pd.DataFrame) and not page_counts_df.empty:
        # 只顯示前幾欄，避免太寬
        cols = list(page_counts_df.columns)
        keep = cols[: min(8, len(cols))]
        show_df = page_counts_df[keep].copy()
        html.append("<div class='card'><table><thead><tr>")
        for c in show_df.columns:
            html.append(f"<th>{escape(str(c))}</th>")
        html.append("</tr></thead><tbody>")
        for _, r in show_df.iterrows():
            html.append("<tr>" + "".join(f"<td>{escape(str(v))}</td>" for v in r.values) + "</tr>")
        html.append("</tbody></table><div class='muted'>※ 表示は前 8 欄のみ。Excel で完全版をご確認ください。</div></div>")
    else:
        html.append("<div class='card muted'>なし</div>")
    html.append("</div>")

    # Query keywords
    html.append("<div class='sec'><h2>主クエリ一覧</h2>")
    if isinstance(qk_list, list) and qk_list:
        html.append("<div class='card'><ul>")
        for k in qk_list:
            html.append(f"<li><code>{escape(str(k))}</code></li>")
        html.append("</ul></div>")
    else:
        html.append("<div class='card muted'>なし</div>")
    html.append("</div>")

    html.append("<div class='sec muted'>Generated at " + escape(ts) + "</div>")
    html.append("</body></html>")
    return "".join(html)

html_text = render_html(
    outline_df=df_outline,
    faqs_list=faqs,
    cta_text=cta,
    ilinks_list=internal_linking,
    notes_list=notes,
    top_sug_df=sheets.get("top_suggests"),
    page_counts_df=sheets.get("per_page_query_counts"),
    qk_list=QUERY_KEYWORDS,
)

with open(html_path, "w", encoding="utf-8") as f:
    f.write(html_text)

# 也一起保存原始 JSON，方便溯源
with open(json_path, "w", encoding="utf-8") as f:
    json.dump({"outline_data": outline_data, "user_brief": USER_BRIEF_SAFE}, f, ensure_ascii=False, indent=2)

print(f"✅ 已輸出 HTML：{html_path}")
print(f"✅ 已輸出 JSON：{json_path}")
````
